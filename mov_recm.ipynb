## **Movie Recomendation Model**

```python
df.head()
```

```python
moviemat = df.pivot_table(index="user_id", columns ="title", values="rating")
moviemat.head()
```

```python
ratings.head()
```

```python
ratings.sort_values('num of ratings',ascending=False)
```

```python
starwars_user_ratings = moviemat['Star Wars (1977)']

starwars_user_ratings.head()
```

```python
similar_to_starwars = moviemat.corrwith(starwars_user_ratings)
similar_to_starwars
```

```python
corr_starwars = pd.DataFrame(similar_to_starwars,columns = ['Corelations'])
corr_starwars.dropna(inplace=True)
```

```python
corr_starwars.head()
```

```python
corr_starwars.sort_values('Corelations',ascending=False).head(10) #you can say top 10 movies
```

```python
ratings
```

```python
corr_starwars
```

```python
corr_starwars = corr_starwars.join(ratings['num of ratings'])
```

```python
corr_starwars.head()
```

```python
corr_starwars['num of ratings'] >100
```

```python
# recomended movies :)
corr_starwars[corr_starwars['num of ratings'] >100].sort_values("Corelations",ascending=False)
```

## **Predict Function**

```python
def predict_movies(movie_name):
    movie_user_ratings = moviemat[movie_name]
    similar_to_movie = moviemat.corrwith(movie_user_ratings)
    
    corr_movie = pd.DataFrame(similar_to_movie,columns = ['Corelations'])
    corr_movie.dropna(inplace=True)
    corr_movie = corr_movie.join(ratings['num of ratings'])
    predictions = corr_movie[corr_movie['num of ratings'] >100].sort_values("Corelations", ascending=False)
#     predictions = predictions.head()
    return predictions
    

    
```

```python
predictions = predict_movies('Star Wars (1977)')
response = predictions.head()
response = list(response.index)
for movie in response:
    print(movie)
```
